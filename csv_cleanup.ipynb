```python
def clean(row):
    if len(re.findall('[^\x00-\x7F]',row))==0:
        return(row)
    else:
        return(re.sub('[^\x00-\x7F]', '', row))     
```

```python
for col in df_original.columns:
    if df_original[col].dtype == ('O'):
        df_original[col] = df_original[col].apply(lambda x: clean(str(x)))
```

```python
def detect_ascii(row):
    row = row.astype('str')
    lst = re.findall('[^\x00-\x7F]',''.join(row.values.tolist()))
    if len(lst)!=0:
        return(True)
    else:
        return(False)
```

```python
df_original.loc[df_original.apply(lambda x: detect_ascii(x),axis = 1),'non_ascii'] = 'Y'
```
