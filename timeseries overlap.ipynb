```python
import pandas as pd
```

```python
csv = r"C:\Users\Shanjeev\Downloads\excel.csv"

df  = pd.read_csv(csv)
```

```python
df['new_start']  = pd.to_datetime(df['new_start'],format="%Y-%m-%d %H:%M:%S")

df['new_end']  = pd.to_datetime(df['new_end'],format="%Y-%m-%d %H:%M:%S")
```

```python
%%time
for x in range(df.shape[0]-1):
    if (df.loc[x,'new_end'] >= df.loc[x+1,'new_start']) & (df.loc[x,'new_end'] <= df.loc[x+1,'new_end']) & (df.loc[x,'id']== df.loc[x+1,'id']):
        df.loc[x+1,'new_start'] = df.loc[x,'new_start']
        df.loc[x,'flag'] = True
    elif (df.loc[x,'new_end'] >= df.loc[x+1,'new_start']) & (df.loc[x,'new_end'] >= df.loc[x+1,'new_end']) & (df.loc[x,'id']== df.loc[x+1,'id']):
        df.loc[x+1,'new_start'] = df.loc[x,'new_start']
        df.loc[x+1,'new_end'] = df.loc[x,'new_end']
        df.loc[x,'flag'] = True
```

```python
final_df = df[df['flag']!=True]

final_df.reset_index(inplace=True)

final_df = final_df.drop(columns = ['flag','index'])
```

```python
df.start = df.new_start.shift(1)

df.end = df.new_end.shift(1)

df = df.assign(previous_id = df.id.shift(1))
```

```python
def stabilzer(row):
    if (row['new_end'] >= row['start']) & (row['new_end'] <= row['end']) & (row['id'] == row['previous_id']):
        row['new_start'] = row['start']
```

```python
df.apply(lambda x: stabilzer(x),axis=1)
```
